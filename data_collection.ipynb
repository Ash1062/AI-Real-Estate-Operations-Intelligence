```python
from google_play_scraper import reviews, Sort
import pandas as pd

app_id = "com.appfolio.resident_portal"

result, continuation_token = reviews(
    app_id,
    lang="en",
    country="us",
    sort=Sort.NEWEST,
    count=500
)

df = pd.DataFrame(result)
df.head()
```
